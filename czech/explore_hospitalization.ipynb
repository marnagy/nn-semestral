```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv('hospitalization.csv')
```

```python
for col in df.columns:
    if col in ['id', 'datum']:
        continue
        
    df[col].fillna(0, inplace=True)
    df[col] = df[col].astype(np.int8)

df.head()
```

```python
df.to_csv('hospitalization.csv', index=False)
```
